```python
import csv
import nltk
from nltk.corpus import words as nltk_words

nltk.download('words')

def is_english_word(word):
    return word.lower() in nltk_words.words()

def count_non_english_words(csv_file):
    non_english_word_count = 0
    with open(csv_file, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            lyrics = row['Lyrics']
            # Split lyrics into words and check each word
            for word in lyrics.split():
                if not is_english_word(word):
                    print(word)
                    non_english_word_count += 1
    return non_english_word_count

def main():
    csv_file = '/Users/prathamgupta/Desktop/trainingLabeled.csv'  # Replace 'your_csv_file.csv' with your CSV file path
    non_english_word_count = count_non_english_words(csv_file)
    print("Number of non-English words in the CSV file:", non_english_word_count)

if __name__ == "__main__":
    main()
```
